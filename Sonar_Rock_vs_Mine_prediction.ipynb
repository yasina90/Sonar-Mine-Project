# Sonar Rock vs Mine Prediction with Python

import the Dependencies



```python
# import libraries

import numpy as numpy
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Data Processing

```python
# loading dataset
sonar_data = pd.read_csv("/content/sonar.csv", header= None)
```

```python
# numbers of raws and columns
sonar_data.shape
```

```python
sonar_data.describe() #Describe --> Statistical measures of the data
```

```python
sonar_data[60].value_counts()
```

M --> Mine

R --> Rock



```python
sonar_data.groupby(60).mean()
```

### Seperate data and lables


```python
# Seperate data and lables

X = sonar_data.drop(columns= 60, axis= 1)
Y =sonar_data[60]
```

```python
X
```

```python
Y
```

### Tain Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size =0.1, stratify =Y, random_state =110 )
```

```python
print("X_train: ",X_train,"\nshape: ", X_train.shape, "X_test", "\n", X_test,"\nshape: ", X_test.shape )
```

```python
print("Y_train: ",Y_train,"\nshape: ", Y_train.shape, "Y_test", "\n", Y_test,"\nshape: ", Y_test.shape )
```

### Model Training ==> Logistic Regression

```python
model = LogisticRegression()
```

### Training the logistic regression model

```python
model.fit(X_train, Y_train)
```

### Evaulaion of Model

```python
# Evaulation with accuracy
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score( y_pred= X_train_prediction, y_true= Y_train)
```

```python
print("training_data_accuracy: ", training_data_accuracy)
```

### Evaulation with accuracy


```python
# Evaulation with accuracy
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(y_pred= X_test_prediction, y_true= Y_test )
```

```python
print("test_data_accuracy: ", test_data_accuracy)
```
